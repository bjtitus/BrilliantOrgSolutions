# Bringing Balance to the Mobile

In this problem we are trying to balance rods by using the Center of Mass equations.

`CM` is the Center of Mass. Set to 0 to solve for other values

`D` is the position relative to the Center of Mass

`W` is the weight of point

\begin{equation*}
CM = {\sum (W \times D) \over \sum W}
\end{equation*}

```python
from sympy import symbols, solve
```

```python
L = symbols('L')
x = symbols('x')
```

```python
def isBalanced(expr):
    sol = solve(expr, dict=True)
    if not sol:
        print("✅ Balanced!")
    else:
        print("❌ Unbalanced")
```

```python
# Rod 1 has 2 & 4
rod1 = (1*(-2*L)+2*L)/(2+1)
isBalanced(rod1)
```

```python
# Rod 2 has 1 & 2
rod2 = (1*(-2*L)+2*L)/(2+1)
isBalanced(rod1)
```

```python
# Rod 3 has 1 & 5
rod3 = (1*(-3*L)+5*L)/(5+1)
isBalanced(rod3)
```

```python
rod3Weight1 = (1*(-3*L)+5*L+1*x)/(5+1+1)
solve(rod3Weight1, x)
```

```python
rod3Weight2 = (1*(-3*L)+5*L+2*x)/(5+1+2)
solve(rod3Weight2, x)
```

```python
# Rod 4 has 6 and 3
rod4 = (6*-2*L+3*3*L)/(6+3)
isBalanced(rod4)
```

```python
rod4weight2 = (6*-2*L+3*3*L+3*x)/(6+3+3)
solve(rod4weight2, x)
```

```python
# Rod 4 (the top) with the adjusted 8 and 12 - the 1 can be ignored
rod5 = (8*-3*L+12*2*L)/(8+12)
isBalanced(rod5)
```
